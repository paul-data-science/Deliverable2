```python
import geopy
import pandas as pd

from tqdm import tqdm
tqdm.pandas()

from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="njit-app")

from geopy.extra.rate_limiter import RateLimiter
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)

nyc_crime['location'] =nyc_crime.progress_apply(lambda row: reverse((row['latitude'], row['longitude'])),axis=1)

def parse_zipcode(location):
    if location and location.raw.get('address') and location.raw['address'].get('postcode'):
        return location.raw['address']['postcode']
    else:
        return None
nyc_crime['zipcode'] = nyc_crime['location'].apply(parse_zipcode)
```
